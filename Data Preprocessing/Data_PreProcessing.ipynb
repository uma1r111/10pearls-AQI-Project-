```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, RobustScaler
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load dataset
df = pd.read_csv("/content/karachi_weather_apr1_to_current.csv")
```

```python
# feature distribution plot
features_to_plot = df.columns.drop("datetime") if "datetime" in df.columns else df.columns
for col in features_to_plot:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```

```python
# check skewness (for log transform)
skewness = df[features_to_plot].skew().sort_values(ascending=False)
print(skewness)

# POSTIVE SKEWNESS
# A positive skewness value (greater than 0) indicates that the right tail of the distribution
# is longer or heavier than the left tail. This often means the mean is greater than the median,
# and the median is greater than the mode. Outliers tend to be on the right side of the distribution.

# NEGATIVE SKEWNESS
# A negative skewness value (less than 0) indicates that the left tail of the distribution is longer or heavier
# than the right tail. This often means the mean is less than the median, and the median is less than the mode.
# Outliers tend to be on the left side of the distribution
```

```python
# Apply log1p to selected skewed features
log_features = ["co", "pm2_5", "pm10", "precip_mm", "so2", "no2"]
for col in log_features:
    df[f"log_{col}"] = np.log1p(df[col])
```

```python
# Scale features using StandardScaler
scaler = StandardScaler()
scaled_features = ["temp_C", "humidity_%", "windspeed_kph", "log_pm2_5", "log_pm10", "log_precip_mm",
                   "log_co", "log_no2", "log_so2", "o3"]
scaled_df = pd.DataFrame(scaler.fit_transform(df[scaled_features]), columns=[f"scaled_{col}" for col in scaled_features])
```

```python
# Combine scaled features with target
df_scaled = pd.concat([scaled_df, df["aqi_us"]], axis=1)
```

```python
print("Preprocessing complete. Scaled dataset preview:")
display(df_scaled.head())
```

```python
final_df = pd.concat([df, df_scaled], axis=1)
final_df.to_csv("full_preprocessed_aqi_weather_data.csv", index=False)
print("Saved full dataset with original + transformed + scaled features.")
```

```python
# Load dataset
df = pd.read_csv("/content/full_preprocessed_aqi_weather_data_with_all_features.csv")
```

```python
df.info()
```

```python
df.index.name
df.index[:5]
```

```python
# Ensure datetime is processed and set as index
df['datetime'] = pd.to_datetime(df['datetime'], dayfirst=True)
df.set_index('datetime', inplace=True)
```

```python
# Create time-based features
df['hour'] = df.index.hour
df['day_of_week'] = df.index.dayofweek  # 0 = Monday, 6 = Sunday
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)  # 1 for Sat/Sun
```

```python
# Interaction terms
df['log_pm2_5_scaled_log_windspeed_kph'] = df['log_pm2_5'] * df['scaled_log_windspeed_kph']
df['scaled_temp_C_scaled_o3'] = df['scaled_temp_C'] * df['scaled_o3']
df['scaled_temp_C_scaled_log_windspeed_kph'] = df['scaled_temp_C'] * df['scaled_log_windspeed_kph']
```

```python
# Save to CSV with datetime index
df.to_csv('full_preprocessed_aqi_weather_data_with_all_features.csv', index=True)
```
